## Intro to BambooLib

[Source](https://bamboolib.8080labs.com/)

**bamboolib** is a GUI for pandas DataFrames.

It provides an intuitive GUI that exports Python code and supports all common transformations and visualizations.
Bamboolib is integrated with Databricks, allowing for a user-friendly exploration of the DBFS, as well as other useful features.
